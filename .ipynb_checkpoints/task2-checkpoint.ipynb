```python
def summary(df):
    print(f'data shape: {df.shape}')
    summ = pd.DataFrame(df.dtypes, columns=['data type'])
    summ['#missing'] = df.isnull().sum().values * 100
    summ['%missing'] = df.isnull().sum().values / len(df)
    summ['#unique'] = df.nunique().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['min'] = desc['min'].values
    summ['max'] = desc['max'].values
    
    return summ
```

## Features choosing and fit-predict

```python
X = train_features[['cell_number', 'execution_count', 'linesofcode', 'variable_count',
       'function_count', 'text/plain', 'text/html',
       'display_data', 'stream', 'error']]

y = train_features['primary_label']

rfe = RFECV(estimator = XGBClassifier(max_depth = 7, 
                           n_estimators = 500, learning_rate = 0.01), min_features_to_select = 2, cv = 5).fit(X, y)

print(X.columns[rfe.support_])
```

```python
X = train_features[['cell_number', 'execution_count', 'linesofcode', 'variable_count',
       'function_count', 'text/plain', 'image/png', 'text/html',
       'display_data', 'stream', 'error']]
y = train_features['primary_label']

X_val = validation_features[['cell_number', 'execution_count', 'linesofcode', 'variable_count',
       'function_count', 'text/plain', 'image/png', 'text/html',
       'display_data', 'stream', 'error']]
y_val = validation_features['primary_label']

test = test_features[['cell_number', 'execution_count', 'linesofcode', 'variable_count',
       'function_count', 'text/plain', 'image/png', 'text/html',
       'display_data', 'stream', 'error']]
# y = np.log(y)
```

### CatBoostClassifier

```python
cat_model = CatBoostClassifier()

cat_parameters = {'learning_rate': [0.03, 0.1],
                  'iterations': [50, 100, 300, 500, 1000],
                  'depth': [4, 6, 10], 
                  'l2_leaf_reg': [0.5, 1, 3, 5, 7, 9],
}

cat_grid = RandomizedSearchCV(estimator=cat_model, 
                             param_distributions = cat_parameters,
                             cv = 5, 
                             n_jobs=5,)

cat_grid.fit(X, y, silent=True)
```

```python
cat_parameters_best = cat_grid.best_params_
print(cat_grid.best_params_)
```

```python
cat_model_best = CatBoostRegressor(**cat_grid.best_params_)

cat_model_best.fit(X, y, silent=True)

pred_cat = cat_model_best.predict(test)
```

```python
val_cat = cat_model_best.predict(X_val).round()

f1_score(list(y_val), list(val_cat), average='weighted')
```

### XGBClassifier

```python
xgb_model = XGBClassifier()

xgb_parameters = {'nthread':[4],
                  'learning_rate': [0.05, 0.1],
                  'max_depth': [4, 6, 8],
                  'min_child_weight': [5, 11, 20],
                  'n_estimators': [50, 100, 300, 500, 1000],
                  'min_samples_leaf': [1, 2, 4],
                  'min_samples_split': [2, 5, 10],
}


xgb_grid = RandomizedSearchCV(estimator=xgb_model, 
                             param_distributions = xgb_parameters,
                             cv = 5, 
                             n_jobs=5)

xgb_grid.fit(X, y)
```

```python
xgb_parameters_best = xgb_grid.best_params_
print(xgb_grid.best_params_)
```

```python
xgb_model_best = XGBRegressor(**xgb_parameters_best)

xgb_model_best.fit(X, y)

pred_xgb = xgb_model_best.predict(test)
```

```python
val_xgb = xgb_model_best.predict(X_val).round()

f1_score(list(y_val), list(val_xgb), average='weighted')
```
